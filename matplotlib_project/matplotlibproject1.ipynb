```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Polygon, Arc

fig, ax = plt.subplots()

# Face
face = Circle((6, 6), 5, color="#FDB813")
ax.add_patch(face)

# Sunglasses
glass = Polygon([[1,8],[5,8],[4.7,6],[1.3,6]], color="black")
glass2 = Polygon([[7,8],[11,8],[10.7,6],[7.3,6]], color="black")
bridge = Polygon([[5,8],[7,8],[6.7,6],[5.3,6]], color="#FDB813")

ax.add_patch(glass)
ax.add_patch(glass2)
ax.add_patch(bridge)

# Side screws
ax.add_patch(Circle((1.4,7),0.25,color="gray"))
ax.add_patch(Circle((10.6,7),0.25,color="gray"))

# Smile
smile = Arc((6,4.5),3.5,2.5,theta1=200,theta2=340,
            linewidth=3,color="saddlebrown")
ax.add_patch(smile)

ax.set_xlim(0,12)
ax.set_ylim(0,12)
ax.set_aspect("equal")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import numpy as np

fig, ax = plt.subplots(figsize=(6,6))

# Petals
petal_radius = 0.12
distance = 0.22

for angle in np.linspace(0, 2*np.pi, 8, endpoint=False):
    x = 0.5 + distance * np.cos(angle)
    y = 0.5 + distance * np.sin(angle)
    petal = Circle((x, y), petal_radius, color='hotpink')
    ax.add_patch(petal)

# Center
center = Circle((0.5, 0.5), 0.14, color='gold')
ax.add_patch(center)

# Stem
ax.plot([0.5, 0.5], [0.0, 0.21], color='green', linewidth=6)

# Leaves
ax.fill([0.5, 0.62, 0.5], [0.01, 0.16, 0.10], color='limegreen')
ax.fill([0.5, 0.38, 0.5], [0.01, 0.16, 0.10], color='limegreen')

ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_aspect('equal')
ax.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc, Ellipse
import numpy as np

fig, ax = plt.subplots(figsize=(6,6))

# Background
ax.set_facecolor("#87CEEB")

# Face
face = Circle((0.5, 0.5), 0.38,
              facecolor="#FFD93D",
              edgecolor="orange",
              linewidth=4)
ax.add_patch(face)

# Eyes
for x in [0.38, 0.62]:
    eye = Circle((x, 0.62), 0.055, color="black")
    ax.add_patch(eye)

    # Eye shine
    ax.add_patch(Circle((x-0.015, 0.64), 0.015, color="white"))
    ax.add_patch(Circle((x+0.012, 0.60), 0.006, color="white"))

# Eyebrows
ax.plot([0.31,0.43],[0.73,0.76],color="black",lw=3)
ax.plot([0.57,0.69],[0.76,0.73],color="black",lw=3)

# Blush
ax.add_patch(Circle((0.26,0.45),0.04,color="pink",alpha=0.5))
ax.add_patch(Circle((0.74,0.45),0.04,color="pink",alpha=0.5))

# Mouth
mouth = Arc((0.5,0.42),0.35,0.28,
            theta1=200,theta2=340,
            lw=5,color="black")
ax.add_patch(mouth)


# Settings
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_aspect("equal")
ax.axis("off")

plt.show()
```
